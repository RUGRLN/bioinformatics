#Determining the protein concentration with correction for light scatter

To determine the protein concentration of a sample, a wavelength scan must be performed of the protein sample using a spectrophotometer over a range of minimum of 200 to 400 nm. Prior to this, the spectrophotometer instrument must be blanked / baseline set using the buffer solution in which the protein has been eluted into.

The protein concentration is ultimately determined at A<sub>280</sub> (apparent absorbance reading at 280 nm).

However, scattering due to light can lead to an inaccurate calculation, which must be corrected for. This is performed by taking multiple absorbance readings over the wavelength range 320 to 350 nm.
<br/>

To correct for light scattering, a graph must be ploted over the wavelength range 320 to 350 nm, using equation (1):


$$ \log{(A_{s, \lambda})} = m\times\log{\lambda} + c \quad \quad (1)$$ 

where A<sub>s,λ</sub> refers to the absorbance due to light scattering at the respective wavelength of λ.

### Wavelength scan for example protein (PduA*)

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pylab as pylab
pylab.rcParams['figure.figsize'] = 10,6  # that's default image size for this interactive session

#Importing CSV file containing the absorbance values versus corresponding wavelength values
data = np.genfromtxt('PduAstar_19112014.CSV', delimiter=',')
wavelength = data[1:,0]
absorbance = data[1:,1]

#Plotting a graph of the data using matplotlib
plt.plot(wavelength, absorbance)
plt.xlabel('Wavelength (nm)')
plt.ylabel('Absorbance (AU)')
plt.grid()
plt.title('Absorbance vs. wavelength for PduA* in buffer A')
plt.savefig('foo.png')
```

###Plotting equation 1

```python
import math 
import numpy as np


#Determining the index of 1-D array 'wavelength' for where 320 and 350 nm lies
a = np.where(wavelength==350)
b = np.where(wavelength==320)
start = int(a[0])
end = int(b[0])

X = [math.log(value, 10) for value in wavelength[start:end]]
Y = [math.log(value, 10) for value in absorbance[start:end]]

p = np.polyfit(X,Y,1)
gradient = p[0]
c = p[1]

fig = plt.figure()
ax1 = fig.add_subplot(111)
ax2 = ax1.twiny()

Xaxis = X
Yaxis = Y

ax1.plot(X,Y)
ax1.grid()
ax1.set_xlabel(r"$\log(\lambda)$", fontsize=18)
ax1.set_ylabel("$\log(A_{s,\lambda})$", fontsize=18)
ax1.set_title("Logarithm of abosrbance due to scattering versus logarithm of wavelength \n \n", fontsize=14)

ax1.text(X[-1],Y[-1],"$y=%.3f x + %.3f$" % (gradient, c), fontsize=16)
ax1.plot(X,np.polyval(p,X), label="Interpolated data", color='blue')
ax1.plot(X, Y,color='red', label="Raw data")
ax1.legend()

ax2.set_xlabel(r"Wavelength (nm)")

a = ax1.get_xticks().tolist()
ax2.set_xlim([a[1], max(a)])

b = ["%.0f" % (10**value) for value in a]
ax2.set_xticklabels(b)
```

```python
#Extrapolating from the graph, to calculate the absorbance due to light scattering at 280 nm
log_280 = math.log(280,10)
log_A_s280 = np.polyval(p, log_280)
A_s280 = 10**log_A_s280

from IPython.display import Latex
Latex("$$A_{(s, 280)} = %f \quad AU$$" % A_s280)
```

##Calculating the theoretical molar extinction coefficient for given protein

```python
from Bio_Seq import *
protein = parse_FASTA("PduA-star.fasta")

seq = protein.values()
n_Trp = 0
n_Tyr = 0
n_Cys = 0
n_SS = 0
for res in seq[0]:
    if res == "W":
        n_Trp += 1
    elif res == "Y":
        n_Tyr += 1
    elif res == "C":
        n_Cys += 1
 
pairs = (n_Cys//2)
n_SS += pairs

ext_coef = ((n_Trp*5500)+(n_Tyr*1490)+(n_SS*125))
              
Latex("Number of tryptophan residues:$$ %.0f $$ Number of tyrosine residues:$$ %.0f $$ Number of disulphide bonds: $$%0.f$$ Molar extinction coefficient: $$\epsilon = %0.f \quad Mcm^{-1}$$" % (n_Trp, n_Tyr, n_SS, ext_coef))
```

```python
#Calculating the molecular weight of the protein
mw = prot_mr(seq[0])/1000
Latex("Molecular weight of protein: $$%f \quad kDa$$" % mw)
```

The protein concentration in mol/dm<sup>3</sup> is calculated according to the following equation (2):
$$[c] = \frac{A_{280}}{\epsilon\times l} \quad (2)$$

Using the moles equation (3), the protein concentration in mg/mL can be calculated using equation #4:

$$moles=\frac{mass}{M_{R}} \quad (3) \\$$

$$[c] = \frac{A_{280}\times M_{R}}{\epsilon\times l} \quad (4)$$

```python
#Calculating the protein concentration in mg/mL
A_280 = absorbance[np.where(wavelength==280)] - A_s280
c = (A_280* mw*1000) / (ext_coef)
Latex("Protein concentration in mg/mL for 3.5 mL sample: $$%.2f \quad mg/mL$$" % c)
```

Protein concentration for a 1 mL sample:

```python
c_f = (c * 3.5)/1
Latex("Protein concentration in mg/mL for 1 mL sample: $$%.2f \quad mg/mL$$" % c_f)
```
